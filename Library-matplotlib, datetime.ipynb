# matplotlib

matplotlib은 그래프를 그릴 때 사용하는 라이브러리입니다.

### plt.scatter()

`plt.scatter(x, y)`는 산포도를 그릴 때 사용합니다. `x`, `y`에는 리스트, 배열 등이 들어갑니다. 예를 들어 아래와 같이 입력하면, (1, 1) (2, 4) (3, 9) 세 개의 좌표에 점이 찍힌 걸 볼 수 있습니다.

```python
# 모듈 임포트
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3], [1, 4, 9])  
plt.show()                         # 그래프를 보여달라는 의미의 명령어. 
```

색상 지정, 모양 변경 등 다양한 옵션을 줄 수 있습니다.

```python
plt.scatter([1, 2, 3], [1, 4, 9],
           color='#ff00ff', # 색 지정
           marker='^',    # 점의 모양
           alpha=0.5)     # 투명도 설정  
plt.show()           
```

### plt.plot()
`plt.plot()`은 입력한 좌표를 선으로 이은 그래프를 그려줍니다.

```python
plt.plot([1, 2, 3], [1, 4, 9])
plt.show()                         
```

이를 활용해 사인 함수를 그려보겠습니다. `np.linspace`를 활용합니다. (넘파이 튜토리얼 참조)

```python
import numpy as np

x = np.linspace(1, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

마찬가지로 다양한 옵션을 줄 수 있습니다.

```python
plt.plot(x, y, 
        color='g',      # 초록색으로 색 변경
        linestyle='--', # 스타일 변경
        linewidth=12)   # 선 굵기 변경
plt.show()
```

여러 그래프를 겹쳐서 그릴 수도 있습니다.

```python
plt.plot(x, y, 
        color='g',      # 초록색으로 색 변경
        linestyle='--', # 스타일 변경
        linewidth=12)   # 선 굵기 변경

plt.plot([1, 10], [0, 1]) 
plt.scatter([0, 10], [-1, 0])
plt.show()
```

### plt.legend()

`plt.legend()`는 범례를 보여줍니다.

```python
plt.plot(x, y, 
        color='g',      
        linestyle='--', 
        linewidth=12,
        label='sin')  

plt.plot([1, 10], [0, 1], label='line') 
plt.scatter([0, 10], [-1, 0], label='dot')

plt.legend()
plt.show()
```

### plt.xlim(), plt.ylim()

` plt.xlim()`, `plt.ylim()`을 통해, 보여지는 화면의 범위를 조절할 수 있습니다.

```python
plt.plot(x, y, 
        color='g',      # 초록색으로 색 변경
        linestyle='--', # 스타일 변경
        linewidth=12)   # 선 굵기 변경

plt.xlim([1, 8]) # x축에 대해 [1, 8] 범위에서만 보이도록 만듭니다.
plt.ylim([-1, 1])
plt.show()
```

### plt.hist()
`plt.hist()`은 히스토그램을 그려줍니다.

```python
x = np.random.randn(1000) # 정규분포에서 임의 추출
plt.hist(x,               # 히스토그램 그리기
         bins=10,         # 막대의 개수
         density=True)    # 밀도 함수로 그릴 것인지 여부
plt.show()
```

# datetime
`datetime`은 시간을 다룰 때 사용합니다. datetime 객체는 아래와 같이 생성합니다.

```python
from datetime import datetime, timedelta

# 연, 월, 일, 시, 분, 초를 적어줍니다. 시, 분, 초는 생략할 수 있습니다.
dt = datetime(2018, 12, 26, 12, 27, 20)
dt
```

```python
# 다음과 같은 방식으로 datetime 객체의 연, 월, 일, 시, 분, 초에 접근할 수 있습니다.
print(dt.year)
print(dt.month)
print(dt.day)
print(dt.hour)
```

```python
# datetime 객체 간 크기 비교가 가능합니다. 
datetime(2019, 1, 1) > datetime(2018, 1, 1)
```

### .weekday()
`.weekday()`를 이용하면 해당 날짜의 요일을 알 수 있습니다. 월요일부터 차례대로 0~6의 수가 부여됩니다.

```python
print(dt.weekday())
```

### datetime.today()

`datetime.today()`는 오늘의 datetime 객체를 가져옵니다.

```python
datetime.today()
```

### .replace()
`.replace()`를 이용해 datetime 객체의 시간을 변경할 수 있습니다.

```python
print('변경 전:', dt)
print('변경 후:', dt.replace(year=dt.year - 2)) # 2년 전으로 변경
```

### strftime : datetime 객체를 str로
`strftime`를 이용하면 datetime 객체를 문자열로 바꿀 수 있습니다.

`%m`은 월, `%d`은 일, `%y`은 연도 등으로 포멧팅해서 원하는 방식의 문자열로 바꿀 수 있습니다.

```python
dt.strftime('%m %d %y // %H %M')
```

### strptime : str을 datetime으로
거꾸로 `strptime`을 이용하면 문자열을 datetime 객체로 바꿀 수 있습니다. 문자열에서 숫자가 있는 부분을 포멧에 맞게 지정해주면 됩니다. 

어떤 포멧을 지원하는지는 다음 링크에서 확인하세요. https://docs.python.org/ko/3.7/library/datetime.html#strftime-and-strptime-behavior

```python
print(datetime.strptime('18 /aa1225', '%y /aa%m%d'))
```

### timedelta()
datetime 객체는 특정한 시간을 나타냅니다. 반면 timedelta 객체는 시간의 길이를 나타냅니다. 예를 들어, 2019년 1월 1일에서 2018년 1월 1일을 빼면, 365일이라는 길이가 나옵니다.

```python
datetime(2019, 1, 1) - datetime(2018, 1, 1)
```

datetime 객체와 timedelta 간 연산을 할 수 있습니다. 예를 들어 다음 코드는 내일의 datetime 객체를 가져옵니다.

```python
datetime.today() + timedelta(2)
```

```python
from datetime import date
```

```python
from datetime import date
date(2019, 12, 1)
```
