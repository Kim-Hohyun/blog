# MNIST 손글씨 데이터 생성하기

필요한 모듈을 불러오고 MNIST 데이터를 로드합니다. 테스트 데이터는 사용하지 않았습니다.

```python
from torch import nn, optim
import torch 
from torch.utils.data import DataLoader
import torchvision.datasets as data
from torchvision.transforms import transforms

train = data.MNIST("./", train=True, transform=transforms.ToTensor(), target_transform=None, download=True)
train_loader = DataLoader(train, batch_size=256, shuffle=64)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

생성자 모델을 만듭니다. 생성자는 표준정규분포를 따르는 100개의 숫자를 받아서 MNIST 이미지 크기인 (1, 28, 28) 크기의 텐서를 만듭니다.

입력되는 값 z는 (batch size, 100, 1, 1)이 될 것입니다.

z가 생성자를 거치면 (batch size, 1, 28, 28)이 됩니다.

z는 1x1 크기의 픽셀 이미지의 필터가 100개 있는 것과 같습니다.

일반적인 convolutional layer와 다르게 transposed convolution을 이용하면 입력되는 이미지의 크기를 키울 수 있습니다.

마지막의 활성화층은 sigmoid 함수를 써서 0\~1 사이의 값으로 만들어줍니다. 이는 입력되는 이미지 값이 0\~1 사이의 분포를 갖기 때문입니다.

구분자는 실제 MNIST 이미지와 생성자가 만든 이미지를 구분해야 합니다.

```python
n_noise = 100

# 중복되는 부분은 함수로 만들어서 간결하게 만들었습니다.
def transposed(in_channels, out_channels, kernel_size=4, stride=2, padding=1):
  return nn.Sequential(
          nn.ConvTranspose2d(in_channels, out_channels, kernel_size, stride, padding),
          nn.BatchNorm2d(out_channels),
          nn.ReLU()
      )

generator = nn.Sequential(
    transposed(n_noise, 256, kernel_size=4, stride=1, padding=0), # (256, 4, 4)
    transposed(256, 128),                                         # (128, 8, 8)
    transposed(128, 32, padding=2),                               # (32, 14, 14)
    nn.ConvTranspose2d(32, 1, kernel_size=4, stride=2, padding=1),# (1, 28, 28)
    nn.Sigmoid()
)
```

다음은 구분자 모델을 정의합니다. 이미지 분류 문제에서 흔히 하듯이 convolutional layer를 이용해 특징을 추출하고 마지막에 완전연결 층을 추가해줍니다.

```python
# discriminator

# 중복되는 부분은 함수로 만들어서 간결하게 만들었습니다.
def conv(in_channels, out_channels, kernel_size=3, stride=1, padding=1):
  return nn.Sequential(
      nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding),
      nn.BatchNorm2d(out_channels),
      nn.ReLU(),
      nn.Dropout2d(0.4),
  )

conv_layer = nn.Sequential(
    conv(1, 128),
    conv(128, 128),
    nn.MaxPool2d(2, 2),        # (14, 14)
    
    conv(128, 128, padding=0),  # (12, 12)
    conv(128, 128),
    nn.MaxPool2d(2, 2),        # (6, 6)
    
    conv(128, 128),
    conv(128, 128),
    nn.MaxPool2d(2, 2),        # (3, 3)
    
    conv(128, 128, padding=0), # (1, 1)
)   
  
mlp_layer = nn.Sequential(
    nn.Linear(128, 200),
    nn.BatchNorm1d(200),
    nn.ReLU(),
    nn.Linear(200, 1)
)

class FlattenLayer(nn.Module):
  def forward(self, x):
    batch_size = x.size(0)
    return x.view(batch_size, -1)

discriminator = nn.Sequential(conv_layer, FlattenLayer(), mlp_layer)
```

훈련시키기

```python
# GPU에서 훈련시켰습니다.
generator.to(device)
discriminator.to(device)

# 생성자와 구분자를 따로 훈련시킬 것입니다.
g_optim = optim.Adam(generator.parameters())
d_optim = optim.Adam(discriminator.parameters())

loss_fn = nn.BCEWithLogitsLoss()
```

구분자는 실제 이미지는 1, 가짜 이미지는 0으로 구분해야 합니다. 반면 생성자는 구분자가 가짜 이미지를 1로 구분하도록 만들어야 합니다. 

```python
batch_size = 256

# 정답 레이블로 사용할 텐서 정의
ones = torch.ones(batch_size).to(device)
zeros = torch.zeros(batch_size).to(device)

# 훈련 중간에 생성자가 어떤 이미지를 생성하는지 확인하기 위한 고정된 입력값
fixed_z = torch.randn(1, n_noise, 1, 1).to(device)
```

```python
# BatchNorm layer는 훈련할 때와 평가할 때를 달리 지정해줘야 한다.
generator.train()
discriminator.train()

generated_imgs = []

for epoch in range(100):
  print("epoch %d" %(epoch+1))
  for i, (real_img, _) in enumerate(train_loader):
    if i % 50 == 0:
      print(">", end='')
    
    real_img = real_img.to(device)
    batch_size = len(real_img)
    
    # 생성자 모델 훈련
    z = torch.randn(batch_size, n_noise, 1, 1).to(device)
    fake_img = generator(z)
    fake_img_detached = fake_img.detach() # 구분자 모델 훈련을 대비한 가짜 이미지
    out = discriminator(fake_img)
    generator_loss = loss_fn(out.squeeze(), ones[:batch_size])
    generator_losses.append(generator_loss.item())

    g_optim.zero_grad(), d_optim.zero_grad()
    generator_loss.backward()
    g_optim.step()
    
    # 구분자 모델 훈련. 가짜 이미지 손실값 + 진짜 이미지 손실값의 그레디언트를 구해서 업데이트를 한다.
    out = discriminator(fake_img_detached)
    discriminator_loss = loss_fn(out.squeeze(), zeros[:batch_size])
    out = discriminator(real_img)
    discriminator_loss += loss_fn(out.squeeze(), ones[:batch_size])
    discriminator_losses.append(discriminator_loss.item())

    g_optim.zero_grad(), d_optim.zero_grad()
    discriminator_loss.backward()
    d_optim.step()
    
  print()
  # 생성된 이미지 중간 저장 
  if (epoch+1) % 10 == 0:
    generated_img = generator(fixed_z)
    save_image(generated_img.squeeze(0), "fake%d.jpg" %((epoch+1)//10))
```

이제 훈련된 생성자 네트워크가 얼마나 비슷한 이미지를 생성하는지 확인해보겠습니다.

생성자 네트워크에 입력되는 숫자는 표준정규분포를 따르는 100개의 숫자입니다. 100개의 숫자 중에서 두 개의 숫자만 변경해가면서 생성자 모델에 넣어줬을 때, 생성되는 이미지가 어떻게 변하는지 확인해봤습니다.



```python
def numpy_img(generated_img):
  generated_img = generated_img.squeeze().cpu().detach().numpy()
  plt.imshow(generated_img, cmap='gray')
  plt.axis("off")
z = torch.randn(100).to(device)

plt.figure(figsize=(12, 12))
plt.subplots_adjust(wspace=0, hspace=0)
for i, z0 in enumerate(torch.linspace(-4, 4, 16)):
  for j, z1 in enumerate(torch.linspace(-4, 4, 16), 1):
    # 4, 5번째 수를 바꿔주면서 생성되는 이미지가 어떻게 변하는지 확인
    z[3] = z0 
    z[4] = z1
    generated_img = generator(z.view(1, 100, 1, 1))
    plt.subplot(16, 16, i*16+j)
    numpy_img(generated_img)
    
plt.show()
```
